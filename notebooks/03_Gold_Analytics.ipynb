```python
from pyspark.sql.functions import *

silver_df = spark.table(
    "student_success.silver.student_performance"
)
#At Risk Students
gold_at_risk_students = silver_df.filter(
    col("risk_status") == "At Risk"
)
gold_at_risk_students.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "student_success.gold.at_risk_students"
    )
# Grade Summary
gold_grade_summary = silver_df.groupBy(
    "final_grade"
).agg(
    count("*").alias("student_count"),
    round(avg("overall_score"), 2).alias("avg_score"),
    round(avg("attendance_percentage"), 2).alias("avg_attendance")
)
gold_grade_summary.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "student_success.gold.grade_summary"
    )
# School Analysis
gold_school_analysis = silver_df.groupBy(
    "school_type"
).agg(
    count("*").alias("student_count"),
    round(avg("overall_score"), 2).alias("avg_score"),
    round(avg("attendance_percentage"), 2).alias("avg_attendance")
)
gold_school_analysis.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "student_success.gold.school_analysis"
    )
# Study Method Analysis
gold_study_method_analysis = silver_df.groupBy(
    "study_method"
).agg(
    count("*").alias("student_count"),
    round(avg("overall_score"), 2).alias("avg_score"),
    round(avg("attendance_percentage"), 2).alias("avg_attendance")
)
gold_study_method_analysis.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "student_success.gold.study_method_analysis"
    )
    
#display(gold_at_risk_students)
#display(gold_grade_summary)
#display(gold_school_analysis)
display(gold_study_method_analysis)
```
